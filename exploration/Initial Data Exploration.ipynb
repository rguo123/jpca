# Initial Data Exploration
This notebook is for running eda on the example data found in the original jPCA code.

```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import scipy
import pandas
```

## Load the data

```python
from scipy.io import loadmat

x = loadmat("../data/exampleData.mat")
x = x['Data']
print(x.shape)
```

### Data Format:
From the NOTES.pdf:

Original Data is in (1, 108) (array of a list containing 108 elements). Each element in the list is a condition for a reaching exercise that the monkey performed. Each condition is wrapped in a superfluous list. Once you access the actual data in the condition, there are two arrays. The first array is a txn array where t is the number of time steps (61) and n is the number of neurons (218) containing the average firing rate (over trials). The second array seems to record the time index but I'm unsure about that.

## View Data
We will begin by plotting one condition.

```python
condition = x[0][0][0]
time = x[0][0][1]
```

```python
condition = np.array(condition)
```

```python
condition.shape
```

```python
time = np.array(time)
time = time.flatten()
```

```python
plt.plot(time, condition[:, 0])
plt.title("Average Neuron 0 Firing Rate during Reaching")
plt.xlabel("Time (ms)")
plt.ylabel("Neural Response")
plt.show()
```

```python
plt.plot(time, condition[:, 50])
plt.title("Average Neuron 50 Firing Rate during Reaching")
plt.xlabel("Time (ms)")
plt.ylabel("Neural Response")
plt.show()
```

```python
for i in range(0, 200, 10):
    plt.plot(time, condition[:, i])
plt.title("Average Neuron Firing Rate during Reaching")
plt.xlabel("Time (ms)")
plt.ylabel("Neural Response")
plt.show()
```
